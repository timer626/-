#### Задача 3. Функция задана таблицей. Применяя метод наименьших квадратов, приблизить функцией вида $\Phi(x)=a_0 \varphi_0(x)+a_1\varphi_1(x)$. Определить среднеквадратичное отклонение. На одном чертеже построить точечный график функции и графики функции $\Phi(x)$.

**Дано** :
$ \normalsize \varphi_0(x)=1 \qquad \varphi_1(x)=(x-1)^3$

|   | 0 | 1 | 2 | 3 | 4 | 5 |
|---|---|---|---|---|---|---|
| x | 0.7 | 0.8 | 3 | 4.2 | 5.9 | 6.6 |
| y | 3.057 | 3.087 | 15.9 | 55.529 | 191.338 | 284.086 |

$\large n=5$

**Метод наименьших квадратов для приближения функции многочленами выглядит следующим образом :**
<tex>$$\large \sum_{j=0}^m \bigg( \sum_{i=0}^n\varphi_j(x_i)\varphi_k(x_i)\bigg)a_j = \sum_{i=0}^n y_i\varphi_k(x_i) ; \quad{k=0,1,...m} \quad (1) \quad \normalsize \text{ где m-количество базисных функий,  n - количество  экспериментов (нумерация начинается с 0)} $$ </tex>

**Среднеквадратичное отклонение :**
<tex>$$\large \delta(Ф,y)=\sqrt{\large \frac{1}{n+1}\large \sum_{i=0}^n (Ф(x_i)-y_i)^2}\quad (2) \quad \normalsize \text{n - количество  экспериментов (нумерация начинается с 0)} $$</tex>

**Вычисление функции $\Phi(x)$ :**

Формула (1) принимает вид: <tex>$$\large \sum_{j=0}^2 \bigg( \sum_{i=0}^5\varphi_j(x_i)\varphi_k(x_i)\bigg)a_j = \sum_{i=0}^5 y_i\varphi_k(x_i) ; \quad{k=0,1} $$</tex>
\begin{equation*}
\begin{cases}
\large \bigg( \sum_{i=0}^5\varphi_0(x_i)\varphi_0(x_i)\bigg)a_0 + \bigg( \sum_{i=0}^5\varphi_1(x_i)\varphi_0(x_i)\bigg)a_1 = \sum_{i=0}^5 y_i\varphi_0(x_i)\\
\large \bigg( \sum_{i=0}^5\varphi_0(x_i)\varphi_1(x_i)\bigg)a_0 + \bigg( \sum_{i=0}^5\varphi_1(x_i)\varphi_1(x_i)\bigg)a_1 = \sum_{i=0}^5 y_i\varphi_1(x_i)
\end{cases}
\Rightarrow \begin{cases} \small 6a_0+((0.7-1)^3+(0.8-1)^3+(3-1)^3+(4.2-1)^3+(5.9-1)^3+(6.6-1)^3)a_1=3.057+3.087+15.9+55.529+191.338+284.086
\\
                                  ((0.7-1)^3+(0.8-1)^3+(3-1)^3+(4.2-1)^3+(5.9-1)^3+(6.6-1)^3)a_0+((0.7-1)^6+(0.8-1)^6+(3-1)^6+(4.2-1)^6+(5.9-1)^6+(6.6-1)^6)a_1=3.057\cdot(0.7-1)^3 +3.087\cdot(0.8-1)^3+15.9\cdot(3-1)^3+55.529\cdot(4.2-1)^3+191.338\cdot(5.9-1)^3+284.086\cdot(6.6-1)^3
                    \end{cases}
                    \large \Rightarrow \begin {cases} 6a_0+334a_1= 553\\
                                               334a_0+45820a_1=74347
                    \end {cases}
\large \Rightarrow a_0=3.1,a_1=1.6
\end{equation*}

$\large \Phi(x)=3.1+1.6(x-1)^3$

Формула (2) принимает вид:
    <tex>$$\large \delta(Ф,y)=\sqrt{\large \frac{1}{6}\large \sum_{i=0}^5 (Ф(x_i)-y_i)^2}$$</tex>
$ \delta(Ф,y)=\sqrt { \frac{1}{6}\big[(3.1+1.6(0.7-1)^3-3.057)^2+(3.1+1.6(0.8-1)^3-3.087)^2+(3.1+1.6(3-1)^3-15.9)^2+(3.1+1.6(4.2-1)^3-55.529)^2+(3.1+1.6(5.9-1)^3-191.338)^2+(3.1+1.6(6.6-1)^3-284.086)^2=0.0003}$

**Ответ:**

$\large \Phi(x)=3.1+1.6(x-1)^3 \quad \delta=0.0003$

```python
import numpy as np
import matplotlib.pyplot as plt

# Определение функции Phi(x)
def Phi(x):
    return 3.1 + 1.6 * (x - 1)**3

# Заданные точки
x_values = [0.7, 0.8, 3, 4.2, 5.9, 6.6]
y_values = [3.057, 3.087, 15.9, 55.529, 191.338, 284.086]

fig, ax = plt.subplots(figsize=(8, 8))
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Построение графика точек
plt.plot(x_values, y_values, 'bo-', linewidth=1, markersize=4, label='Точки')

# Построение графика функции Phi(x)
x = np.linspace(-10, 10, 100)
y = Phi(x)
plt.plot(x, y, 'r-', linewidth=2, label='Функция Ф(x)')


plt.grid(linestyle='--', alpha=0.7)

# Изменение границ осей
plt.xlim(-10, 10)
plt.ylim(-300, 300)

# Установка делений осей с промежутками длиной в 1
major_ticks = np.arange(-300, 350, 50)
major_ticks_1 = np.arange(-10, 11, 1)
ax.set_xticks(major_ticks_1)
ax.set_yticks(major_ticks)
plt.title('График', fontsize=16)

# Улучшение внешнего вида делений
ax.tick_params(axis='both', direction='in', length=6, width=1, color='gray')

# Добавление стрелок на концы осей
arrow_format = dict(facecolor='black', edgecolor='black', linewidth=1)
plt.annotate('', xy=(10, 0), xytext=(9.8, 0), arrowprops=dict(arrowstyle='->', **arrow_format,
                                                          shrinkA=0, shrinkB=0))
plt.annotate('', xy=(0, 10), xytext=(0, 9.8), arrowprops=dict(arrowstyle='->', **arrow_format,
                                                          shrinkA=0, shrinkB=0))

# Добавление названий осей
plt.text(9.8, -10, 'X', fontsize=14, ha='right', va='center', color='gray')
plt.text(-0.5, 10, 'Y', fontsize=14, ha='center', va='bottom', color='gray')

# Добавление легенды
plt.legend(loc='upper left')

plt.tight_layout()
plt.show()
```
