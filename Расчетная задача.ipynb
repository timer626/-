### Задача 1. Решить переопределенную систему  и найти среднеквадратичную ошибку : $\normalsize\|Ax-b\|_{\Large e}$
##### Дано:
$\large Ax=b$
\begin{equation*}
\begin{aligned}
& \large A = 
\begin{pmatrix}
2 & -8\\
5 & -2\\
-5 & 2\\
2 & 7\\
-7 & 7
\end{pmatrix}
&
& \large b = 
\begin{pmatrix}
-34.3\\
-31.2\\
31\\
11\\
56
\end{pmatrix}
\end{aligned}
\end{equation*}
#### Решение :
Переопределенная система имеет вид : $\large A^{T}A=A^{T}b$, где $\large A^{T}=\begin{pmatrix}
2 & 5 & -5 & 2 & -7\\
-8 & -2 & 2 & 7 & 7
\end{pmatrix}$  
Пусть:
\begin{equation*}
\begin{cases}
\large \tilde {A} &= \large A^{T}A\\
\large \tilde b &= \large A^{T}b
\end{cases}
\large \Rightarrow \large \tilde{A}x= \tilde{b}
\end{equation*}
Найдем $\large \tilde{A} :$
$\large A^{T}A=\begin{pmatrix} \large
2 & 5 & -5 & 2 & -7\\
-8 & -2 & 2 & 7 & 7
\end{pmatrix}
*
\begin{pmatrix}
2 & -8\\
5 & -2\\
-5 & 2\\
2 & 7\\
-7 & 7
\end{pmatrix}
= \begin{pmatrix}
107 & -71\\
-71 & 170
\end{pmatrix}$

Найдем $\large \tilde{b}: $
$\large A^{T}b=\begin{pmatrix} \large
2 & 5 & -5 & 2 & -7\\
-8 & -2 & 2 & 7 & 7
\end{pmatrix}
*
\begin{pmatrix}
-34.3\\
-31.2\\
31\\
11\\
56
\end{pmatrix}
= \begin{pmatrix}
-749.6\\
867.8
\end{pmatrix}$

Решение системы :
$\large \overline{x} =
\begin{pmatrix}
-5.006\\
3.014\\
\end{pmatrix}
$


Рассмотрим выражение 
$\large A\overline{x} - b:$

$
\large A\overline{x}-b= \begin{pmatrix}
2 & -8\\
5 & -2\\
-5 & 2\\
2 & 7\\
-7 & 7
\end{pmatrix}*
\begin{pmatrix}
-5.006\\
3.014\\
\end{pmatrix}
-\begin{pmatrix}
-34.3\\
-31.2\\
31\\
11\\
56
\end{pmatrix}
=\begin{pmatrix}
0.176\\
0.142\\
0.058\\
0.086\\
0.14
\end{pmatrix}
$    


Вычислим $\normalsize\|Ax-b\|_{\Large e} :$

$\normalsize\|Ax-b\|_{\Large e}= \sqrt{0.176^2+0.142^2+0.058^2+0.086^2+0.140^2}=0.286$

Ответ: $\large \overline{x} =
\begin{pmatrix}
-5.006\\
3.014\\
\end{pmatrix}
,
\normalsize\|Ax-b\|_{\Large e}= 0.286
$
